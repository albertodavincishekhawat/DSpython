## How to use **countplot**?

> sns works with palette only no indivisual colors

> it's made using matplotlib

> in matplotlib columns are to be provided, while using sns we don't need to provide 

```python
plt.title("Count plot")
plt.xlabel('Gender')
plt.ylabel('Count')
sns.countplot(df, x='gender', palette="mako") # palette = "rocket"
plt.show()
```

```python
plt.title("Count plot")
plt.xlabel('Gender')
plt.ylabel('Count')
sns.countplot(df, x='gender', palette="mako", hue = "day") # palette = "rocket"   hue =
plt.show()
```

## How to make **KDEplot**?
key density estimator

> hue can be catagorical data

```python
# plot
plt.title("KDEplot")
sns.kdeplot(df, x = "total_bill", fill = True, hue = "time", )
```

```python
# plot
plt.title("KDEplot")
sns.histplot(df, x = "total_bill", hue = "time", kde=True ) #fill = True (no need to use fill here)
```

```python
# plot
plt.title("KDEplot")
sns.kdeplot(df, x = "total_bill", hue = "day", fill = True )
```

```python
plt.figure(figsize=(8,8))
sns.jointplot(df, x="total_bill", y = "tip", palette = "mako", )
```

```python
plt.figure(figsize=(8,8))
sns.jointplot(df, x="total_bill", y = "tip", palette = "mako", hue = "gender")
```

```python
# data
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.jointplot(iris, x="petal_length", y = "petal_width", hue = 'species', kind = 'kde')
```
